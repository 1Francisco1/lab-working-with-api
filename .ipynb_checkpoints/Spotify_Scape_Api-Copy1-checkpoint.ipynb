```python
# It is limited to 100 tracks, we will have to fix it:

def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks

tracks = get_playlist_tracks("spotify", "7dUHWQzNHsfzQNuPkg0scx")
```

```python
# Getting all the artists IDs

def get_artists_and_ids_from_playlist(playlist_id):
    
    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)
    
    
    artists_ids = []
    
    for track in tracks_from_playlist:
        artists_info = track['track']['artists']
        
        for artist_info in artists_info:
            artists_ids.append(artist_info['id'])
            
            
            
    return list(set(artists_ids))
```

```python
# Getting all the artists

def get_artists_from_playlist(playlist_id):
    
    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)
    
    artists = []
    
    for track in tracks_from_playlist:
        artist_info = track['track']['artists'][0]
        artists.append(artist_info['name'])
    
    return artists
```

```python
#getting all track ids
track_ids = [track['track']['id'] for track in tracks]

#getting song names from track ids
song_actual_names = [track['track']['name'] for track in tracks]

#get artist (from function before)
artist_list = get_artists_from_playlist("7dUHWQzNHsfzQNuPkg0scx")
```

```python
#get all genres and popularity

def get_pop_and_genre():
    
    genre = []
    popularity = []
    
    for elem in tracks:
        artist01 = sp.artist(elem['track']["album"]["artists"][0]["external_urls"]["spotify"])
        genre.append(artist01["genres"][0] if artist01["genres"] else "generless")
        popularity.append(artist01["popularity"])
        
        
    return genre,popularity

genre,popularity = get_pop_and_genre()
```

```python
#having the specs of each song, and checking where they are (in a nested list in this case)
liest = []

for i in track_ids:
    liest.append(sp.audio_features(i))
    
liest
```

```python
#looping inside the nested list in order to have the entire track list able to be converted into a df
var_name = [i for elem in liest for i in elem]
```

```python
#converting entire track list into df
import pandas as pd
df01 = pd.json_normalize(var_name)
```

```python
#creating the dataframes to append later.

df = pd.DataFrame(song_actual_names, columns = ["song_name"])
df_artists = pd.DataFrame(artist_list,columns = ["artists"])
df_genre = pd.DataFrame(genre,columns = ["genre"])
df_popularity = pd.DataFrame(popularity,columns = ["popularity"])
```

```python
maybe = pd.concat([df, df01,df_artists,df_genre,df_popularity], axis=1)
maybe
```

```python
pd.set_option('display.max_columns', None)
maybe.head()
```

Treating the dataframe:

```python
from sklearn.preprocessing import StandardScaler

df_to_scale = maybe[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','popularity']]

scaler = StandardScaler()

scaled_df = scaler.fit_transform(df_to_scale)
```

```python
final_df = pd.DataFrame(scaled_df, columns=df_to_scale.columns)
final_df["song_name"] = maybe["song_name"]
final_df["artists"] = maybe["artists"]
final_df["genre"] = maybe["genre"]
final_df
```

```python
#since there are too many, silhouette will be used to check better clustering number
final_df.genre.nunique()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn import datasets
from matplotlib import pyplot as plt
import numpy as np

#applying silhouette to see how many clusters is a good number to apply

K = range(2, 20)

silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    silhouette.append(silhouette_score(scaled_df, kmeans.predict(scaled_df)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_df)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(scaled_df)
clusters
```

```python
final_df["clusters"] = clusters
```

```python
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
final_df.groupby(['clusters','genre'], as_index=False).count().sort_values(['clusters', 'key'], ascending=[True, False])[['genre', 'clusters', 'key']].reset_index(drop=True).groupby(["clusters"]).head(3)
```

```python
import sklearn
from sklearn.metrics import pairwise_distances_argmin_min
#getting song and recomending song kinda here
def recomend_song():
    song_name = input("Choose a song: ")
    results = sp.search(q=f'track:{song_name}', limit=1)
    track_id = results['tracks']['items'][0]['id'] if results['tracks']['items'] else None
    while not track_id:
        return recomend_song()
    # get song features with the obtained id
    audio_features = sp.audio_features(track_id)
    #get song popularity and genre
    song_pop_genre = sp.artist(results['tracks']["items"][0]["album"]["artists"][0]["external_urls"]["spotify"])
    the_genre = song_pop_genre["genres"][0] if song_pop_genre["genres"] else "generless"
    the_pop = song_pop_genre["popularity"]
    #create the df
    df_uni_song = pd.DataFrame(audio_features)
    df_uni_song["popularity"] = the_pop
    #merge the datas
    new_features = df_uni_song[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','popularity']]
    df_uni_song["genre"] = the_genre
    #scale features
    scaled_x = scaler.transform(new_features)
    
    # predict cluster
    cluster = kmeans.predict(scaled_x)
    # filter dataset to predicted cluster
    filtered_df = final_df[final_df['clusters'] == cluster[0]][new_features.columns]
    # get closest song from filtered dataset
    closest, _ = pairwise_distances_argmin_min(scaled_x, filtered_df)
    # return it in a readable way
    print('\n [RECOMMENDED SONG]')
    return ' - '.join([final_df.loc[closest]['song_name'].values[0], final_df.loc[closest]['artists'].values[0]])
```
